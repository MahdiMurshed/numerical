```python
def f(x):
    return x**10-1
```

```python
x=np.arange(1,100,1)
y=f(x)
```

```python
plt.plot(x,y)
```

```python
def bisec(x0,x1,N,e):
    step=1
    condition=True
    flag=1
    while condition:
        x2=(x0+x1)/2
        print('\nIteration-%d : x0= %f x1= %f x2=%f f(x2)= %0.6f e=%f '%(step,x0,x1,x2,f(x2),abs(f(x2))))
        if(f(x0)*f(x2)<0):
            x1=x2
        else:
            x0=x2
        step=step+1
        condition=abs(f(x2))>e

    
    print('\nRequired root is : %0.6f '% x2)
        
        
```

```python
def falsePosition(x0,x1,N,e):
    step=1
    condition=True
    flag=1
    while condition:
        x2=x0-f(x0)*(x1-x0)/(f(x1)-f(x0))
        print('\nIteration-%d : x0= %f x1= %f x2=%f f(x2)= %0.6f e=%f '%(step,x0,x1,x2,f(x2),abs(f(x2))))
        if f(x0)*f(x2)<0:
            x1=x2
        else:
            x0=x2
        step=step+1
        condition=abs(f(x2))>e
    print('\nRequired root is : %0.6f '% x2)
    
        
```

```python
x0=float(input('first guess : '))
x1=float(input('second guess : '))
N=int(input('Step : '))
e=float(input('e : '))
if(f(x0)*f(x1)>0):
    print('Wrong guess')
else:
    print('\nBisection method')
    bisec(x0,x1,N,e)
    print('\nFalse Position')
    falsePosition(x0,x1,N,e)
    
```

# Open methods

```python
def f(x):
    return np.exp(-x)-x
```

Fixed Point

```python
def g(x):
    return np.exp(-x)
```

```python
def fixedPoint(x0,e):
    step=1
    condition=True
    while condition:
        x2=g(x0)
        print('\nIteration-%d : x0= %f  x2=%f f(x2)= %0.6f e=%f '%(step,x0,x2,f(x2),abs(f(x2))))
        x0=x2
        step=step+1
        if step>N:
            condition=False
    print('\nRequired root is : %f'%x2)
        
    
    
    
```

Newton Rhapson

```python
def fp(x):
    return -(np.exp(-x)+1)
```

```python
def NR(x0,e):
    step=1
    condition=True
    while condition:
        if fp(x0)==0:
            print('\nerror')
        x2=x0-f(x0)/fp(x0)
        print('\nIteration-%d : x0= %f  x2=%f f(x2)= %0.6f e=%f '%(step,x0,x2,f(x2),abs(f(x2))))
        x0=x2
        step=step+1
        if step>N:
            condition=False
    print('\nRequired root is : %f'%x2)
         
        
```

Secant Method

```python
def secant(x0,x1,e):
    step=1
    condition=True
    while condition:
        if f(x0) == f(x1):
            print('Divide by zero error!')
            break
        x2=x0-f(x0)*(x1-x0)/(f(x1)-f(x0))
        print('\nIteration-%d : x0= %f  x2=%f f(x2)= %0.6f e=%f '%(step,x0,x2,f(x2),abs(f(x2))))
        x0=x1
        x1=x2
        step=step+1
        if step>N:
            condition=False
    print('\nRequired root is : %f'%x2)
        
```
